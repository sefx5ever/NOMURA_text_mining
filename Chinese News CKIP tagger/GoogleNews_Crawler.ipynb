```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import requests
from fake_useragent import UserAgent
from urllib.parse import urlparse
from urllib import parse
import os
import datetime
```

```python

df = pd.DataFrame(columns = ["Title","YYYY","MM","DD","URL","Content"])
df2 = pd.DataFrame(columns = ["Title","YYYY","MM","DD","URL","Content"])
search_year = 2008

while search_year <= 2021:
    # num: 一次最多要request的筆數, 可減少切換頁面的動作
    url = "https://www.google.com/search?num=100&q=" + "美股+" + "before:"+str(search_year) +"&tbm=nws"
    ua = UserAgent()
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--proxy-server=socks5://localhost:9050')
    driver = webdriver.Chrome(chrome_options = chrome_options)
    driver.get(url)
    header = {'user-agent': ua.random }
    
    res = requests.get(url, headers = header)
    res.encoding = 'utf-8'

    if res.status_code == 200:
        text = res.text
        soup = BeautifulSoup(text, "html.parser")
        #print(soup.prettify())
        items = soup.findAll("div", {"class": "g"})
        for item in items:
            # title
            news_title = item.find("h3", {"class": "r"}).find("a").text
            print(news_title)
            # url
            href = item.find("h3", {"class": "r"}).find("a").get("href")
            print(href)
            # content
            news_text = item.find("div", {"class": "st"}).text
            print(news_text)
            # source
            date = item.find("span", {"class": "f nsa fwzPFf"}).text
            print(date)
            if "年" in date and "月" in date:
                ds = date.split('月')[0].split('年')
                yyyy = ds[0]
                mm = ds[1]
                dd = date.split('月')[1].replace('日','')
            else: # today
                yyyy = datetime.date.today().year
                mm =  datetime.date.today().month
                dd = datetime.date.today().day
            if search_year <= 2014:
                df = df.append({
                    "Title": news_title,
                    "YYYY": yyyy,
                    "MM":mm,
                    "DD":dd,
                    "URL": href,
                    "Content": news_text,
                },ignore_index=True)
                file_name = "00before_" + str(search_year) + ".xlsx"
                df.to_excel(file_name,index = False)
            else:
                df2 = df2.append({
                    "Title": news_title,
                    "YYYY": yyyy,
                    "MM":mm,
                    "DD":dd,
                    "URL": href,
                    "Content": news_text,
                },ignore_index=True)
                file_name = "00before_" + str(search_year) + ".xlsx"
                df2.to_excel(file_name,index = False)
            os.system("killall -HUP tor")
            
    search_year += 1
    driver.close()
```
