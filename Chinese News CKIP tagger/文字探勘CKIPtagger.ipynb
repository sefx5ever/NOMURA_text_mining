```
def flatten(input_list):
    output_list = []
    while True:
        if input_list == []:
            break
        for index, i in enumerate(input_list):
            if type(i)== list:
                input_list = i + input_list[index+1:]
                break
            else:
                output_list.append(i)
                input_list.pop(index)
                break
    return output_list
```

```
def count_tf(w_list):
  tf_list = list()
  w_list2 = w_list
  
  for i,w in enumerate(w_list):
    count_tf = 0
    for w2 in w_list2:
      if w in w2:
        count_tf += 1
    tf_list.append(count_tf)
   
  
  return tf_list
#============================================
def count_df(w_list,d_list):
  
  df_list = list()

  for i,w in enumerate(w_list):
    df_count = 0
    for doc in d_list:
      if w in doc:
        df_count += 1
    df_list.append(df_count)
  return df_list
#============================================
import math
def count_tfidf(doc, gram, tf_list, df_list):
  tfidf_list = []

  for i in range(0,len(tf_list),1):
    tfidf_list.append((1 + math.log(tf_list[i]))* math.log(doc/df_list[i]))

  return tfidf_list
```

```
# calculate tf & df

# ckiptagger
'''
data_utils.download_data("./")
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")
'''
# Load model with GPU
ws = WS("./data", disable_cuda=False)
#pos = POS("./data", disable_cuda=False)
#ner = NER("./data", disable_cuda=False)

sentence_list1 = list(table['Title'])
sentence_list2 = list(table['Content'])
sentence_list1.extend(sentence_list2) # 標題內文概要一起斷
doc_len = len(sentence_list1)
# 斷詞
word_list = ws(sentence_list1)
print(word_list)
gram_len = len(word_list)
```

```

flatten_word_list = flatten(word_list)

# 算df
df1 = pd.DataFrame(columns=['詞','TF', 'DF','TF-IDF'])
df1['詞'] = flatten_word_list
df1['TF'] = count_tf(flatten_word_list)
df1['DF'] = count_df(flatten_word_list,sentence_list1)
df1['TF-IDF'] = count_tfidf(doc_len, gram_len, df1['TF'], df1['DF'])
df1
```

```
df1 = df1.drop_duplicates("詞",'first')
df1.to_excel(key + "_word_" + yr +".xlsx", index = False)
df1
```
